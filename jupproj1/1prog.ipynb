* Рассчитать на какой улице больше остановок

```python
import csv
```

```python
streets = {}
```

```python
%time
with open('data-398-2020-09-09.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        if row['Street'] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
```

```python
streets
```

```python
%time
sorted(streets.items(), key=lambda item: item[1], reverse=True)
```
